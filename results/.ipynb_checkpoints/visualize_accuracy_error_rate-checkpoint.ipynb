```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
def find_percent_error_rate(array_1d):
    return((100 - np.array(array_1d)))
```

```python
def get_data(filename):
    data = pd.read_csv(filename, header=None)
    #data = data.head(5)
    data = data.values
    print(data)
    data_header = data[0,:]
    data = np.delete(data, 0, 0)
    data = find_percent_error_rate(data)
    print(data)
    return(data_header, data)
```

```python
data_header, data = get_data('accuracy_HAR1D.csv')

'''
x = data[0,:-3]
original = data[1,:-3]
randomized = data[2,:-3]
ecl = data[3,:-3]
js = data[4,:-3]


'''
x = data_header
original = data[0,:]
randomized = data[1,:]
ecl = data[2,:]
js = data[3,:]
```

```python
#plt.figure(figsize=(8,6))
plt.rcParams.update({'font.size': 10})
plt.figure(num=None, figsize=(8, 6), dpi=120, facecolor='w', edgecolor='k')
plt.plot(x, original,'g-', linewidth=3.0, label="original")
plt.plot(x, randomized, 'r-', linewidth=3.0, label="shuffled")
plt.plot(x, ecl, 'm-o', linewidth=0.75, label="ecl")
plt.plot(x, js, 'b->', linewidth=0.75,label="js")
plt.xlabel('perplexity')
plt.ylabel('error rate')
#plt.title('Accuracy vs Different Perplexity for ISOLET Dataset')
plt.legend()
plt.show()
```
