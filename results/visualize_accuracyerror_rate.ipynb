```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
def get_data(filename):
    data = pd.read_csv(filename, header=None)
    #data = data.head(5)
    data = data.values
    return(data)
```

```python
data = get_data('accuracy.csv')


x = data[0,:]
original = data[1,:]
randomized = data[2,:]
ecl = data[3,:]
js = data[4,:]


'''
x = data[0,:]
original = data[1,:]
randomized = data[2,:]
ecl = data[3,:]
js = data[4,:]
'''
```

```python
#plt.figure(figsize=(8,6))
plt.rcParams.update({'font.size': 10})
plt.figure(num=None, figsize=(8, 6), dpi=120, facecolor='w', edgecolor='k')
plt.plot(x, original,'g-', linewidth=3.0, label="original")
plt.plot(x, randomized, 'r-', linewidth=3.0, label="randomized")
plt.plot(x, ecl, 'm-o', linewidth=0.75, label="ecl")
plt.plot(x, js, 'b->', linewidth=0.75,label="js")
plt.xlabel('perplexity')
plt.ylabel('accuracy')
#plt.title('Accuracy vs Different Perplexity for HAR Dataset')
plt.legend()
plt.show()
```
