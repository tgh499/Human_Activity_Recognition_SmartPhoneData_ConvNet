```python
import pandas as pd
import numpy as np

data = pd.read_csv('accuracy_HAR1D.csv', header=None)
data = data.values

for i,j in enumerate(data):
    if i>0:
        data[i] = 100 - data[i]


print(data)
data = np.delete(data, 1, 0)
data = np.delete(data, 1, 0)
data = np.transpose(data)

text = ""
text +=  "\hline"
text += "\n"
for i,j in enumerate(data):
    for k in j:
        if k == data[i][0]:
            text += str(int(k))
        else:
            text += str(round(k,4))
        if k != data[i][-1]:
            text += " & "
    text += " \\\\ \n"
    text += "\hline"
    text += "\n"

print(text)
```
