```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
import random
```

```python
def show_sample_2d(filename_prefix, rows, columns, sample_index):
    filename = filename_prefix + ".csv"
    data = pd.read_csv(filename, header=None)

    features = data.columns[1:]
    train_data = data[features]
    train_data = train_data.values

    print(train_data.shape)


    arr = np.reshape(train_data[sample_index], (rows,columns))
    plt.imshow(arr, cmap='gray_r')
    #plt.gca().invert_yaxis()
    plt.show()
```

```python
def view_image(dim, imageplot_filename, perplexity):
    if perplexity != None:
        imageplot_filename = imageplot_filename + str(perplexity) + ".csv"
    data = pd.read_csv(imageplot_filename, header=None)
    
    features = data.columns[1:]
    data_features = data[features]
    data_features_np = data_features.values

    arr = np.reshape(data_features_np[sample_index], (dim,dim), 'C')
    plt.imshow(arr, cmap='gray_r',origin = 'upper')
    #plt.gca().invert_yaxis()
    
    save_to_disk_filename = imageplot_filename[:-4] + ".png"
    print(save_to_disk_filename)
    plt.savefig(save_to_disk_filename)
    plt.show()
```

```python
def show_sample_1d(filename_prefix, sample_index, perplexity):
    if perplexity != None:
        filename_prefix = filename_prefix + str(perplexity) + ".csv"
    data = pd.read_csv(filename_prefix, header=None)
    
    features = data.columns[1:]
    data_features = data[features]
    data_features_np = data_features.values
    
    x = [i for i in range(len(data_features_np[sample_index]))]
    
    #plt.figure(num=None, figsize=(8, 6), dpi=120, facecolor='w', edgecolor='k')
    #plt.plot(x, data_features_np[sample_index], color='w', width=5, edgecolor='k')
    plt.plot(x, data_features_np[sample_index], color='k')
    
    save_to_disk_filename = filename_prefix[:-4] + ".png"
    print(save_to_disk_filename)
    plt.savefig(save_to_disk_filename)
    plt.show()
```

```python
sample_index = 2
#show_sample_2d('test_transformed', 22, 28, 1)
#show_sample_2d('test_randomized', 22, 28, 1)
perplexities = [10, 30, 50, 70, 90, 110, 130, 150, 170, 190, 210, 230, 250, 270, 290, 310, 330, 350]
#perplexities = [110]


show_sample_1d('test_transformed.csv', sample_index, None)
show_sample_1d('test_randomized.csv', sample_index, None)

for i in perplexities:
    filename = "test_randomized_js_"
    #show_sample_2d(filename, 22, 28, sample_index)
    show_sample_1d(filename, sample_index, i)
    
for i in perplexities:
    filename = "test_randomized_ecl_"
    #show_sample_2d(filename, 22, 28, sample_index)
    show_sample_1d(filename, sample_index, i)
```
