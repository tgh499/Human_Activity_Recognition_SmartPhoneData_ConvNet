```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
import random
```

```python
def show_sample_2d(filename_prefix, rows, columns, sample_index):
    filename = filename_prefix + ".csv"
    data = pd.read_csv(filename, header=None)

    features = data.columns[1:]
    train_data = data[features]
    train_data = train_data.values

    print(train_data.shape)


    arr = np.reshape(train_data[sample_index], (rows,columns))
    plt.rcParams.update({'font.size': 10})
    plt.figure(num=None, figsize=(8, 6), dpi=120, facecolor='w', edgecolor='k')
    plt.imshow(arr, cmap='gray_r')
    #plt.gca().invert_yaxis()
    plt.show()
```

```python
show_sample_2d('test_transformed', 17, 33, 1)
show_sample_2d('test_randomized', 17, 33, 1)
perplexities = [10, 30, 50, 70, 90, 110, 130, 150, 170, 190, 210, 230, 250, 270, 290, 310, 330, 350]
#perplexities = [110]

for i in perplexities:
    filename = "test_randomized_ecl_" + str(i)
    show_sample_2d(filename, 17, 33, 1)
```
